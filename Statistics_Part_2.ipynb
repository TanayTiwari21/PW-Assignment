#Statistics Part 2

Q1) What is hypothesis testing in statistics?

Ans: Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (hypothesis) about a population parameter.

Q2) What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:

Null Hypothesis (H₀): It is a statement that there is no effect or no difference, and it serves as the default or starting assumption.

Alternative Hypothesis (H₁): It is a statement that contradicts the null hypothesis and represents what the researcher aims to prove.

Q3) What is the significance level in hypothesis testing, and why is it important?

Ans:
The significance level (denoted by α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05, 0.01, or 0.10. It determines the threshold for statistical significance.

Q4) What does a P-value represent in hypothesis testing?

Ans:
The P-value is the probability of observing test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is true.

Q5) How do you interpret the P-value in hypothesis testing?

Ans:

If P-value ≤ α: Reject the null hypothesis (evidence supports the alternative).

If P-value > α: Fail to reject the null hypothesis (not enough evidence).

Q6) What are Type 1 and Type 2 errors in hypothesis testing?

Ans:

Type 1 Error (α): Rejecting the null hypothesis when it is actually true.

Type 2 Error (β): Failing to reject the null hypothesis when the alternative hypothesis is true.

Q7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans:

One-tailed test: Tests for the possibility of the relationship in one direction.

Two-tailed test: Tests for the possibility of the relationship in both directions.

Q8) What is the Z-test, and when is it used in hypothesis testing?

Ans:
A Z-test is used to determine if there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large (n > 30).

Q9) How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans:
Z = (X̄ - μ) / (σ / √n)
Where X̄ = sample mean, μ = population mean, σ = population standard deviation, and n = sample size.
It represents how many standard deviations the sample mean is from the population mean.

Q10) What is the T-distribution, and when should it be used instead of the normal distribution?

Ans:
The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It accounts for the extra uncertainty in small samples.

Q11) What is the difference between a Z-test and a T-test?

Ans:

Z-test: Used when the population standard deviation is known and sample size is large.

T-test: Used when the population standard deviation is unknown and sample size is small.

Q12) What is the T-test, and how is it used in hypothesis testing?

Ans:
A T-test is used to compare the means of two groups or a sample mean to a known value when the population standard deviation is unknown.

Q13) What is the relationship between Z-test and T-test in hypothesis testing?

Ans:
Both are used to test hypotheses about population means. The T-test adjusts for sample size uncertainty, while the Z-test assumes a known standard deviation and large sample size. As sample size increases, the T-distribution approaches the normal distribution.

Q14) What is a confidence interval, and how is it used to interpret statistical results?

Ans:
A confidence interval gives a range of values within which the true population parameter is likely to lie with a certain level of confidence (e.g., 95%). It provides an estimate of uncertainty.

Q15) What is the margin of error, and how does it affect the confidence interval?

Ans:
The margin of error is the range added and subtracted from the sample statistic to create a confidence interval. A larger margin of error means a wider interval, indicating more uncertainty.

Q16) How is Bayes' Theorem used in statistics, and what is its significance?

Ans:
Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It plays a crucial role in statistical inference, especially in Bayesian statistics.

Q17) What is the Chi-square distribution, and when is it used?

Ans:
The Chi-square distribution is used in hypothesis testing for categorical data. It is commonly used in tests of independence and goodness of fit.

Q18) What is the Chi-square goodness of fit test, and how is it applied?

Ans:
The Chi-square goodness of fit test determines whether a sample data matches an expected distribution. It compares observed and expected frequencies in categories.

Q19) What is the F-distribution, and when is it used in hypothesis testing?

Ans:
The F-distribution is used to compare two variances or in analysis of variance (ANOVA) tests. It is used when testing if multiple population means are equal.

Q20) What is an ANOVA test, and what are its assumptions?

Ans:
ANOVA (Analysis of Variance) is used to test differences between three or more group means.
Assumptions:

Independence of observations

Normally distributed groups

Equal variances (homogeneity of variances)

Q21) What are the different types of ANOVA tests?

Ans:

One-way ANOVA: Tests one factor among multiple groups

Two-way ANOVA: Tests two factors simultaneously

Repeated Measures ANOVA: Tests the same subjects under different conditions

Q22) What is the F-test, and how does it relate to hypothesis testing?

Ans:
An F-test is used to compare two variances or test overall significance in regression or ANOVA. It helps determine whether observed group differences are statistically significant.

```python
# Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm
data = np.random.normal(52, 10, 100)
sample_mean = np.mean(data)
pop_mean = 50
pop_std = 10
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(data)))
p = 2 * (1 - norm.cdf(abs(z)))
print("Z-score:", z)
print("P-value:", p)
```

```python
# Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
sample = np.random.normal(100, 10, 50)
z = (np.mean(sample) - 98) / (10 / np.sqrt(len(sample)))
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.plot(x, y)
plt.axvline(z, color='r', linestyle='--')
plt.axvline(-1.96, color='g', linestyle='--')
plt.axvline(1.96, color='g', linestyle='--')
plt.title("Two-tailed Z-test Decision Region")
plt.show()
```

```python
# Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def visualize_errors(mu0, mu1, sigma, n, alpha):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    z_alpha = norm.ppf(1 - alpha)
    cutoff = mu0 + z_alpha * sigma / np.sqrt(n)
    y0 = norm.pdf(x, mu0, sigma/np.sqrt(n))
    y1 = norm.pdf(x, mu1, sigma/np.sqrt(n))
    plt.plot(x, y0, label="H0")
    plt.plot(x, y1, label="H1")
    plt.fill_between(x, 0, y0, where=(x >= cutoff), color='red', alpha=0.5, label="Type I Error")
    plt.fill_between(x, 0, y1, where=(x < cutoff), color='blue', alpha=0.5, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()
visualize_errors(100, 105, 15, 50, 0.05)
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2, n - 1) * stderr
    return (mean - margin, mean + margin)
data = np.random.normal(120, 20, 50)
print("Confidence Interval:", confidence_interval(data))
```

```python
# Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t
data = np.random.normal(130, 25, 60)
n = len(data)
stderr = np.std(data, ddof=1) / np.sqrt(n)
margin_error = t.ppf(0.975, n - 1) * stderr
print("Margin of Error:", margin_error)
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_a, prob_b_given_a, prob_b):
    return (prob_b_given_a * prior_a) / prob_b
prior_a = 0.01
prob_b_given_a = 0.9
prob_b = 0.05
posterior = bayes_theorem(prior_a, prob_b_given_a, prob_b)
print("Posterior Probability:", posterior)
```

```python
# Q16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f
sample1 = np.random.normal(100, 10, 30)
sample2 = np.random.normal(100, 20, 30)
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)
f_stat = var1 / var2
d1 = len(sample1) - 1
d2 = len(sample2) - 1
p_value = 1 - f.cdf(f_stat, d1, d2)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
# Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
from scipy.stats import shapiro, levene
def check_anova_assumptions(groups):
    for i, group in enumerate(groups):
        print(f"Group {i+1} Shapiro p-value:", shapiro(group).pvalue)
    print("Levene test p-value:", levene(*groups).pvalue)
group1 = np.random.normal(70, 5, 20)
group2 = np.random.normal(75, 5, 20)
group3 = np.random.normal(80, 5, 20)
check_anova_assumptions([group1, group2, group3])
```

```python
# Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
df = pd.DataFrame({
    'Score': np.random.normal(75, 10, 60),
    'Method': np.repeat(['A', 'B'], 30),
    'Teacher': np.tile(['X', 'Y', 'Z'], 20)
})
model = ols('Score ~ C(Method) + C(Teacher) + C(Method):C(Teacher)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
sns.boxplot(x='Method', y='Score', hue='Teacher', data=df)
```

```python
# Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2
data = np.random.normal(50, 5, 25)
sample_var = np.var(data, ddof=1)
pop_var = 25
chi2_stat = (len(data) - 1) * sample_var / pop_var
p_val = 2 * min(chi2.cdf(chi2_stat, df=len(data)-1), 1 - chi2.cdf(chi2_stat, df=len(data)-1))
print("Chi2 Statistic:", chi2_stat)
print("P-value:", p_val)
```

```python
# Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f
x = np.random.normal(100, 10, 30)
y = np.random.normal(100, 20, 30)
f_stat = np.var(x, ddof=1) / np.var(y, ddof=1)
dfn = len(x) - 1
dfd = len(y) - 1
p_val = 1 - f.cdf(f_stat, dfn, dfd)
x_vals = np.linspace(0, 4, 500)
plt.plot(x_vals, f.pdf(x_vals, dfn, dfd))
plt.axvline(f_stat, color='r', linestyle='--')
plt.title("F-distribution and Observed Statistic")
plt.show()
print("F-statistic:", f_stat)
print("P-value:", p_val)
```
