Q6).  Explain the concept of generators in Python and how they are defined.

-> A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.

Generators are defined just like function but instead of return we use yield to iterates.

ex:-

def fun():
    
    yield 1            
    yield 2            
    yield 3            

for val in fun():
    print(val)

output:-
1

2

3



Q7). What are the advantages of using generators over regular functions?

->
Memory Efficiency:
Unlike regular functions that return the entire result at once, generators yield values one by one as needed, reducing memory consumption, especially when dealing with large datasets.

Lazy Evaluation:
Generators compute values only when they are requested, which can improve performance, particularly in scenarios where not all values are required.

Concise Code:
Generators often provide a cleaner and more readable way to create iterators compared to traditional loops.

State Management:
Generators maintain their state between calls, allowing you to resume execution from the last yielded point.

Infinite Sequences:
Generators can be used to create infinite sequences of values, unlike regular functions that need to return a complete result.

Cooperative Multitasking:
Generators can be paused and resumed, facilitating cooperative multitasking where different parts of the code can share the processing load.

Q10). What is the difference between map(), reduce(), and filter() functions in Python?

-> The map(), reduce(), and filter() functions are built-in higher-order functions in Python that operate on iterables (like lists, tuples, etc.). They each serve different purposes:

map():
Applies a given function to each item in an iterable and returns a new iterable (map object) containing the results. It transforms each element individually.

filter():
Filters elements from an iterable based on a given function that returns True or False. It returns a new iterable (filter object) containing only the elements for which the function returns True.

reduce():
Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It is part of the functools module and needs to be imported.

```python
#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum

number=[1,23,44,55,34,65,66,21,7,8,923,87]
sum_even_numbers(number)
```

```python
#Q3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares

numbers = [1, 2, 3, 4, 5]
square_numbers(numbers)
```

```python
#Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number ofterms.

def fibonacci(n_terms):
    a, b = 0, 1
    count = 0
    while count < n_terms:
        yield a
        a, b = b, a + b
        count += 1

a=fibonacci(10)
```

```python
'''Q11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
Write a Python program using lambda and map.
'''

orders = [
    (34587, 4, 40.95),
    (98762, 5, 56.80),
    (77226, 3, 32.95),
    (88112, 3, 24.99)
]

result = list(map(
    lambda order: (
        order[0],
        order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
    ),
    orders
))

print(result)

```
