1. What is unsupervised learning in the context of machine learning?

Ans:-
Unsupervised learning is a type of ML where the model finds hidden patterns or structures in data without labeled outputs. Example: clustering, dimensionality reduction.

2. How does K-Means clustering algorithm work?

Ans:-

K-Means clustering works in a simple iterative process to group data into K clusters:

Choose K ‚Üí the number of clusters you want.

Initialize centroids ‚Üí pick K points (randomly or using K-Means++) as the starting cluster centers.

Assign step ‚Üí each data point is assigned to the nearest centroid (using distance, usually Euclidean).

Update step ‚Üí recompute each centroid as the mean of all points assigned to that cluster.

Repeat steps 3 & 4 until centroids stop changing (or change very little).

3. Explain the concept of a dendrogram in hierarchical clustering.

Ans:-

A dendrogram in hierarchical clustering is a tree-like diagram that shows how data points or clusters are merged (in agglomerative) or split (in divisive) step by step.

The leaves (bottom) represent individual data points.

The branches represent clusters formed by merging points or smaller clusters.

The height (y-axis) shows the distance (or dissimilarity) at which clusters are joined.

4. What is the main difference between K-Means and Hierarchical Clustering?

Ans:-

The main difference lies in how clusters are formed:

K-Means ‚Üí Partitional method

Splits the dataset into K clusters directly.

Requires the number of clusters K to be specified in advance.

Works well for large datasets (faster).

Hierarchical Clustering ‚Üí Hierarchical method

Builds a tree structure (dendrogram) showing how clusters merge (agglomerative) or split (divisive).

Does not require K upfront; you can choose the number of clusters by ‚Äúcutting‚Äù the dendrogram.

More computationally expensive (slower for large datasets).

5. What are the advantages of DBSCAN over K-Means?

Ans:-

Here are the key advantages of DBSCAN over K-Means:

No need to specify K ‚Üí DBSCAN automatically finds the number of clusters, while K-Means requires K upfront.

Arbitrary shape detection ‚Üí DBSCAN can detect clusters of any shape (not just spherical), unlike K-Means which assumes circular clusters.

Handles varying sizes and densities better ‚Üí K-Means struggles when clusters differ in size or density, DBSCAN adapts well.

Outlier detection ‚Üí DBSCAN explicitly identifies noise points (outliers), while K-Means forces every point into a cluster.

6. When would you use Silhouette Score in clustering?

Ans:-

You would use the Silhouette Score when you want to evaluate the quality of clustering results without ground truth labels.

It measures how well each data point fits within its own cluster compared to other clusters.

The score ranges from -1 to +1:

+1 ‚Üí points are well matched to their own cluster and far from others.

0 ‚Üí points are on the boundary between clusters.

Negative ‚Üí points may be in the wrong cluster.

typically use it to:

Compare different clustering algorithms (e.g., K-Means vs DBSCAN).

Choose the best number of clusters (K) in K-Means or hierarchical clustering.

7. What are the limitations of Hierarchical Clustering?

Ans:-

Computationally expensive ‚Üí Time complexity is O(n¬≤) (or worse), making it unsuitable for very large datasets.

Greedy nature ‚Üí Once clusters are merged or split, the decision cannot be undone (no backtracking).

Sensitive to noise and outliers ‚Üí A single outlier can distort the entire hierarchy.

Choice of linkage matters ‚Üí Results can vary significantly depending on whether you use single, complete, average, or Ward linkage.

Interpretability issue for large data ‚Üí Dendrograms become complex and hard to interpret when there are thousands of points.

8. Why is feature scaling important in clustering algorithms like K-Means?

Ans:-

K-Means uses distance (usually Euclidean) to assign points to clusters.

If features are on different scales (e.g., age in years vs. income in lakhs), the feature with the larger range will dominate the distance calculation.

This leads to biased clusters, where some features are ignored unintentionally.

Scaling (e.g., Standardization or Min-Max Normalization) puts all features on a comparable scale, so each contributes fairly to clustering.

9. How does DBSCAN identify noise points?

Ans:-

DBSCAN identifies noise points (outliers) using its two parameters:

Eps (Œµ): the neighborhood radius around a point.

MinPts: the minimum number of points required in that neighborhood to be considered dense.

A point is labeled as noise if:

It is not a core point (i.e., it has fewer than MinPts neighbors within Eps).

It is not directly reachable from any core point (so it can‚Äôt be part of a cluster).

11. What is the elbow method in K-Means clustering?

Ans:-

The Elbow Method is a technique to determine the optimal number of clusters (K) in K-Means.

How it works:

Run K-Means for different values of K (e.g., 1 to 10).

Compute the inertia (sum of squared distances to centroids) for each K.

Plot K vs. Inertia.

The curve will typically decrease rapidly at first, then level off.

The ‚Äúelbow point‚Äù (where the decrease slows) represents the best trade-off between fewer clusters and lower inertia.


12. Describe the concept of "density" in DBSCAN.

Ans:-

In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the concept of density is central to how clusters are formed.

Density is defined using two parameters:

Eps (Œµ): The radius of a neighborhood around a point.

MinPts: The minimum number of points required within that radius to consider the region dense.

Key ideas:

A point is a core point if it has at least MinPts neighbors within distance Eps ‚Üí this indicates a dense region.

A border point has fewer than MinPts neighbors but lies within the neighborhood of a core point.

Points that are neither core nor border are labeled as noise (outliers).

13. Can hierarchical clustering be used on categorical data?

Ans:-

Yes , hierarchical clustering can be used on categorical data, but not directly with the usual Euclidean distance.

For categorical data, you need to use appropriate similarity/distance measures such as:

Hamming distance ‚Üí proportion of mismatched attributes.

Jaccard similarity/distance ‚Üí good for binary/categorical data.

Simple matching coefficient (SMC).

Sometimes categorical variables are also encoded into numeric form (e.g., one-hot encoding) before applying hierarchical clustering.


14. What does a negative Silhouette Score indicate?

Ans:-

A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster.

Here‚Äôs why:

Silhouette Score compares how similar a point is to its own cluster (cohesion) vs. the nearest other cluster (separation).

A negative value means the point is closer on average to another cluster than to the one it was assigned.

Interpretation:

+1 ‚Üí point is well-clustered.

0 ‚Üí point lies on the boundary between clusters.

Negative ‚Üí point is probably in the wrong cluster (misclassified).

15. Explain the term "linkage criteria" in hierarchical clustering.

Ans:-

In hierarchical clustering, the linkage criteria define how the distance between two clusters is calculated when merging them.

The most common linkage methods are:

Single Linkage (Nearest Neighbor):
Distance between two clusters = minimum distance between any two points (one from each cluster).

Tends to form ‚Äúchain-like‚Äù clusters.

Complete Linkage (Farthest Neighbor):
Distance = maximum distance between any two points from the clusters.

Produces compact, evenly sized clusters.

Average Linkage:
Distance = average distance between all pairs of points (one from each cluster).

Balances single and complete linkage.

Ward‚Äôs Method:
Merges clusters that result in the minimum increase in total variance (within-cluster sum of squares).

Tends to create clusters of similar size.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans:-

K-Means clustering can perform poorly on data with varying cluster sizes or densities because of its assumptions and limitations:

Assumes spherical clusters ‚Üí K-Means works best when clusters are roughly circular. If clusters are elongated or irregular, it misclassifies points.

Assumes equal cluster sizes ‚Üí It tries to balance cluster sizes, so larger clusters may get split, and smaller clusters may get merged.

Sensitive to density differences ‚Üí Dense clusters and sparse clusters are treated equally, which causes misgrouping.

Every point must belong to a cluster ‚Üí K-Means does not detect noise/outliers; it forces them into clusters, distorting results.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans:-

The two core parameters in DBSCAN are:

1. Eps (Œµ ‚Äì epsilon)

Defines the radius of the neighborhood around a point.

Controls how close points need to be to be considered neighbors.

Small Œµ ‚Üí many points become noise, clusters may be too fragmented.

Large Œµ ‚Üí clusters may merge, losing finer structure.

2. MinPts (Minimum Points)

Minimum number of points required within Œµ-radius for a point to be considered a core point.

Small MinPts ‚Üí even sparse regions may be considered clusters (risk of noise being clustered).

Large MinPts ‚Üí only dense regions form clusters, may miss smaller clusters.

Together:

A core point has at least MinPts within Œµ.

Border points are within Œµ of a core point but don‚Äôt have enough neighbors to be core themselves.

Anything else = noise.

Rule of thumb:

Choose MinPts ‚â• number of dimensions + 1.

Tune Œµ using a k-distance graph (elbow method for density).

19. What is agglomerative clustering?

Ans:-

Agglomerative clustering is the most common type of hierarchical clustering that works in a bottom-up manner.

How it works:

Start with each data point as its own cluster.

At each step, merge the two closest clusters based on a chosen distance metric (e.g., Euclidean, Manhattan) and linkage criteria (single, complete, average, Ward).

Repeat until all points are merged into one big cluster (or until a stopping condition is met).

The process can be visualized with a dendrogram, where you can ‚Äúcut‚Äù the tree at a chosen level to get the desired number of clusters.

Key points:

No need to predefine K (but you can choose it later by cutting the dendrogram).

More computationally expensive than K-Means (about O(n¬≤)).

Good for small to medium datasets and when you want to see hierarchical structure.

```python
#Q21Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.7, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="x", s=200, label="Centroids")
plt.legend()
plt.show()
```

```python
# Q29 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5.
# Display results as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.7, random_state=42)

# Evaluate silhouette scores
scores = {}
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    scores[k] = silhouette_score(X, labels)

# Plot silhouette scores
plt.bar(scores.keys(), scores.values(), color="skyblue")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K=2 to 5")
plt.show()
```

```python
# Q30 Load the Iris dataset and use hierarchical clustering to group data.
# Plot a dendrogram with average linkage.

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical clustering with average linkage
Z = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode="lastp", p=20, leaf_rotation=45, leaf_font_size=10)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.show()
```

```python
# Q31 Generate synthetic data with overlapping clusters using make_blobs,
# then apply K-Means and visualize with decision boundaries.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate overlapping synthetic data
X, y = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap="viridis")
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="x", s=200, label="Centroids")
plt.legend()
plt.title("K-Means with Decision Boundaries")
plt.show()
```

```python
# Q34 Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means.
# Show results in a line plot.

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for K=2 to 6
inertias = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot inertia values
plt.plot(range(2, 7), inertias, marker="o")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("K-Means Inertia for K=2 to 6")
plt.show()
```

```python
# Q39 Generate synthetic non-linearly separable data using make_moons,
# apply K-Means, and visualize the clustering result.

from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate moon-shaped data
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# Apply KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="x", s=200, label="Centroids")
plt.legend()
plt.title("KMeans on Non-Linearly Separable Data (Moons)")
plt.show()
```

```python
# Q40 Load the Digits dataset, apply PCA to reduce to 3 components,
# then use KMeans and visualize with a 3D scatter plot.

from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce to 3D using PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

# Plot 3D scatter
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels, cmap="tab10", s=15)
ax.set_title("KMeans Clustering on Digits (3D PCA Reduced)")
plt.show()
```

```python
#Q42 Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate data
X, y = make_blobs(n_samples=600, centers=5, cluster_std=0.8, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(X)

# Evaluate
score = silhouette_score(X, labels)
print("Silhouette Score:", score)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="tab10", s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="x", s=200, label="Centroids")
plt.legend()
plt.show()
```

```python
#Q43 Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load dataset
data = load_breast_cancer()
X = data.data

# PCA for 2D
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agg.fit_predict(X_pca)

# Plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="coolwarm", s=30)
plt.title("Agglomerative Clustering on Breast Cancer (PCA 2D)")
plt.show()
```

```python
#Q44 Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate data
X, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap="viridis", s=30)
axes[0].set_title("KMeans Clustering")

axes[1].scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap="viridis", s=30)
axes[1].set_title("DBSCAN Clustering")

plt.show()
```

```python
#Q45 Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X = iris.data

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Silhouette Coefficients
silhouette_vals = silhouette_samples(X, labels)

# Plot
plt.bar(range(len(silhouette_vals)), silhouette_vals, color="skyblue")
plt.title("Silhouette Coefficient for each sample (Iris)")
plt.show()
```

```python
#Q48 Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate data
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters and noise
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print("Number of clusters found:", n_clusters)
print("Number of noise points:", n_noise)
```

```python
#Q49 Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()
X = digits.data

# Reduce dimensions using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_tsne)

# Plot
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap="tab10", s=10)
plt.title("Agglomerative Clustering on Digits (t-SNE 2D)")
plt.show()
```
