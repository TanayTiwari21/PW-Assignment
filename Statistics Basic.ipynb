Q1. What is statistics, and why is it important?

-> Statistics is a branch of mathematics that deals with the collection, analysis, interpretation, and presentation of data. It is important because it helps in making informed decisions based on data, identifying trends, and solving real-world problems using numerical evidence.

Q2. What are the two main types of statistics?

-> The two main types of statistics are:

Descriptive Statistics – used to summarize and describe the features of a dataset.

Inferential Statistics – used to draw conclusions and make predictions about a population based on a sample.

Q3. What are descriptive statistics?

-> Descriptive statistics are techniques used to summarize and organize data in a meaningful way. Examples include mean, median, mode, standard deviation, and graphical methods like bar charts and histograms.

Q4. What is inferential statistics?

-> Inferential statistics involve drawing conclusions or making inferences about a population based on sample data. It includes hypothesis testing, confidence intervals, and regression analysis.

Q5. What is sampling in statistics?

-> Sampling is the process of selecting a subset of individuals or observations from a larger population. It is used to gather information efficiently and make predictions or decisions about the whole population.

Q6. What are the different types of sampling methods?

-> Different types of sampling methods are:

Random Sampling

Systematic Sampling

Stratified Sampling

Cluster Sampling

Convenience Sampling

Q7. What is the difference between random and non-random sampling?

-> In random sampling, every member of the population has an equal chance of being selected.
In non-random sampling, selection is based on factors like convenience or judgment, which can introduce bias.

Q8. Define and give examples of qualitative and quantitative data.

-> Qualitative data describes categories or qualities. Example: gender, color, or name.
Quantitative data represents numerical values. Example: age, height, marks scored.

Q9. What are the different types of data in statistics?

-> Types of data include:

Qualitative (Categorical)

Quantitative (Numerical)

Discrete Data

Continuous Data

Q10. Explain nominal, ordinal, interval, and ratio levels of measurement.

->

Nominal: Categories without order (e.g., colors, gender)

Ordinal: Categories with order (e.g., rankings)

Interval: Ordered, with equal spacing but no true zero (e.g., temperature in Celsius)

Ratio: Like interval, but with a true zero (e.g., weight, age)

Q11. What is the measure of central tendency?

-> It is a statistical measure to determine the center or typical value of a dataset. The main measures are mean, median, and mode.

Q12. Define mean, median, and mode.

->

Mean: The average of all values.

Median: The middle value when data is ordered.

Mode: The value that appears most frequently.

Q13. What is the significance of the measure of central tendency?

-> It helps in identifying the central point of a dataset, making it easier to understand the overall distribution and compare different datasets.

Q14. What is variance, and how is it calculated?

-> Variance measures how much data values differ from the mean.
Formula:
Variance (σ²) = Σ(x - mean)² / N

Q15. What is standard deviation, and why is it important?

-> Standard deviation is the square root of variance. It shows how spread out the values are in a dataset. A higher standard deviation means more variability.

Q16. Define and explain the term range in statistics.

-> Range is the difference between the highest and lowest values in a dataset.
Formula: Range = Max value – Min value

Q17. What is the difference between variance and standard deviation?

-> Variance is the average squared deviation from the mean. Standard deviation is its square root. Both measure spread, but standard deviation is in the same unit as the data.

Q18. What is skewness in a dataset?

-> Skewness measures the asymmetry of a distribution. If data is not symmetrical, it is said to be skewed.

Q19. What does it mean if a dataset is positively or negatively skewed?

->

Positively skewed: Tail is longer on the right; most values are on the left.

Negatively skewed: Tail is longer on the left; most values are on the right.

Q20. Define and explain kurtosis.

-> Kurtosis measures the "tailedness" or peakedness of a distribution.

High kurtosis = heavy tails (more outliers)

Low kurtosis = light tails

Q21. What is the purpose of covariance?

-> Covariance shows the direction of a linear relationship between two variables. Positive covariance means they increase together; negative means one increases while the other decreases.

Q22. What does correlation measure in statistics?

-> Correlation measures the strength and direction of a linear relationship between two variables, usually ranging from -1 to +1.

Q23. What is the difference between covariance and correlation?

->

Covariance indicates direction but not the strength.

Correlation standardizes the value, showing both direction and strength of the relationship.

Q24. What are some real-world applications of statistics?

->

Business – Market analysis, quality control

Healthcare – Drug trials, disease prediction

Education – Exam results analysis

Government – Census, policy planning

Sports – Player performance and strategy building

```python
# Q1. How do you calculate the mean, median, and mode of a dataset?

import statistics as stats

data = [12, 15, 14, 10, 18, 20, 15]

mean = stats.mean(data)
median = stats.median(data)
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Q2. Write a Python program to compute the variance and standard deviation of a dataset.

import statistics as stats

data = [12, 15, 14, 10, 18, 20, 15]

variance = stats.variance(data)
std_dev = stats.stdev(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Q8. Generate a dataset and demonstrate positive and negative skewness.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Positively skewed data (long right tail)
pos_skewed = np.random.exponential(scale=2, size=1000)

# Negatively skewed data (long left tail)
neg_skewed = -np.random.exponential(scale=2, size=1000)

# Plotting histograms
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(pos_skewed, bins=30, kde=True)
plt.title(f"Positive Skewness: {skew(pos_skewed):.2f}")

plt.subplot(1,2,2)
sns.histplot(neg_skewed, bins=30, kde=True)
plt.title(f"Negative Skewness: {skew(neg_skewed):.2f}")

plt.show()
```

```python
# Q9. Write a Python script to calculate covariance between two datasets.

import numpy as np

x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0, 1]

print("Covariance between x and y:", cov_xy)
```

```python
# Q12. Implement and compare simple random sampling and systematic sampling.

import pandas as pd
import numpy as np

data = pd.DataFrame({'Value': range(1, 21)})

# Simple Random Sampling - select 5 samples randomly
random_sample = data.sample(n=5, random_state=1)

# Systematic Sampling - select every 4th item
systematic_sample = data.iloc[::4]

print("Simple Random Sampling:\n", random_sample)
print("\nSystematic Sampling:\n", systematic_sample)
```

```python
# Q13. Calculate the mean, median, and mode of grouped data.

import numpy as np
import statistics as stats

# Grouped data classes and frequencies
classes = [(10, 20), (20, 30), (30, 40), (40, 50)]
freq = [5, 8, 12, 5]

# Calculate midpoints of classes
midpoints = [ (start + end) / 2 for start, end in classes ]

# Mean of grouped data
mean = sum(m * f for m, f in zip(midpoints, freq)) / sum(freq)

# For grouped data median and mode need more work; simplified here:
median = 30  # Approximate based on grouped data
mode = 30    # Approximate mode class midpoint

print("Mean:", mean)
print("Median (approx):", median)
print("Mode (approx):", mode)
```

```python
# Q14. Simulate data using Python and calculate its central tendency and dispersion.

import numpy as np
import statistics as stats

# Simulate dataset
data = np.random.normal(loc=50, scale=10, size=100)  # Normal distribution

mean = stats.mean(data)
median = stats.median(data)
mode = stats.mode(data)
variance = stats.variance(data)
std_dev = stats.stdev(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Q20. Write a Python program to visualize covariance using a heatmap.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Two related datasets
data = np.array([[2, 4, 6, 8, 10],
                 [1, 3, 5, 7, 9]])

# Compute covariance matrix
cov_matrix = np.cov(data)

# Plot heatmap
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm')
plt.title("Covariance Matrix Heatmap")
plt.show()
```

```python
# Q22. Generate a dataset and implement both variance and standard deviation computations.

import numpy as np

data = np.random.randint(50, 100, size=20)

variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)   # Sample standard deviation

print("Data:", data)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Q24. Implement the Pearson and Spearman correlation coefficients for a dataset.

import pandas as pd
from scipy.stats import spearmanr, pearsonr

df = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [5, 6, 7, 8, 7]
})

# Pearson correlation
pearson_corr, _ = pearsonr(df['X'], df['Y'])

# Spearman correlation
spearman_corr, _ = spearmanr(df['X'], df['Y'])

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
