Q2) What are the key assumptions of Simple Linear Regression
Ans:-

Linearity between independent and dependent variables

Independence of residuals

Homoscedasticity (constant variance of residuals)

Normality of residuals

No or little multicollinearity (though only one predictor is involved here)

Q3) What does the coefficient m represent in the equation Y = mX + c
Ans:- The coefficient m is the slope, which indicates the change in Y for a one-unit change in X.

Q4) What does the intercept c represent in the equation Y = mX + c
Ans:- The intercept c is the predicted value of Y when X is 0. It represents where the regression line crosses the Y-axis.

Q6) What is the purpose of the least squares method in Simple Linear Regression
Ans:- The least squares method minimizes the sum of the squared differences between actual and predicted Y values, ensuring the best-fitting line.

Q7) How is the coefficient of determination (R²) interpreted in Simple Linear Regression
Ans:- R² represents the proportion of the variance in the dependent variable that is predictable from the independent variable. Ranges from 0 to 1.

Q8) What is Multiple Linear Regression
Ans:- Multiple Linear Regression is an extension of Simple Linear Regression where the dependent variable is predicted using two or more independent variables.

Q9) What is the main difference between Simple and Multiple Linear Regression
Ans:- Simple Linear Regression uses one independent variable; Multiple Linear Regression uses two or more.

Q10) What are the key assumptions of Multiple Linear Regression
Ans:-

Linearity

Independence of errors

Homoscedasticity

Normality of residuals

No multicollinearity

Q11) What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model
Ans:- Heteroscedasticity refers to non-constant variance of residuals. It can lead to inefficient estimates and affect the reliability of statistical tests.

Q12) How can you improve a Multiple Linear Regression model with high multicollinearity
Ans:-

Remove or combine highly correlated predictors

Use dimensionality reduction (e.g., PCA)

Apply regularization techniques like Ridge or Lasso regression

Q13) What are some common techniques for transforming categorical variables for use in regression models
Ans:-

One-Hot Encoding

Label Encoding

Binary Encoding

Ordinal Encoding (if categories have order)

Q14) What is the role of interaction terms in Multiple Linear Regression
Ans:- Interaction terms capture the combined effect of two or more variables that is not explained by their individual effects.

Q15) How can the interpretation of intercept differ between Simple and Multiple Linear Regression
Ans:-

In Simple Linear Regression, the intercept is the expected value of Y when X = 0.

In Multiple Linear Regression, the intercept is the expected Y when all predictors are 0, which may or may not be meaningful.

Q16) What is the significance of the slope in regression analysis, and how does it affect predictions
Ans:- The slope indicates how much the dependent variable changes with a unit change in the independent variable. It defines the strength and direction of the relationship.

Q17) How does the intercept in a regression model provide context for the relationship between variables
Ans:- The intercept provides a baseline value of the dependent variable, helping to contextualize predictions when all independent variables are 0.

Q18) What are the limitations of using R² as a sole measure of model performance
Ans:-

It does not indicate model accuracy or bias

Cannot detect overfitting

Does not account for the number of predictors

Q19) How would you interpret a large standard error for a regression coefficient
Ans:- A large standard error indicates that the estimate of the coefficient is unstable and may not be significantly different from zero.

Q20) How can heteroscedasticity be identified in residual plots, and why is it important to address it
Ans:- It is identified by a funnel or fan shape in the residual vs. predicted plot. It must be addressed because it violates regression assumptions and distorts inference.

Q21) What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²
Ans:- It suggests that some predictors may not be contributing to the model and may just be adding noise.

Q22) Why is it important to scale variables in Multiple Linear Regression
Ans:- Scaling ensures that all variables contribute equally to the analysis, especially important for regularization methods and interpretation.

Q27) Can polynomial regression be applied to multiple variables
Ans:- Yes, it can be extended to include polynomial terms of multiple variables, resulting in a more complex model.

Q28) What are the limitations of polynomial regression
Ans:-

Prone to overfitting

Interpretation is difficult

Sensitive to outliers

Computationally expensive for high-degree polynomials

Q29) What methods can be used to evaluate model fit when selecting the degree of a polynomial
Ans:-

Cross-validation

Adjusted R²

AIC/BIC

RMSE or MAE on validation set

Q30) Why is visualization important in polynomial regression
Ans:- Visualization helps understand the shape of the curve, detect overfitting, and communicate model behavior clearly.
