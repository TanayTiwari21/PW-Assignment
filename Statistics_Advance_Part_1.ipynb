Q1. What is a random variable in probability theory?

Ans: A random variable is a variable whose value is determined by the outcome of a random experiment. It maps outcomes of a probabilistic process to numerical values.

Q2. What are the types of random variables?

Ans:

Discrete Random Variable: Takes a countable number of distinct values.

Continuous Random Variable: Takes an infinite number of possible values within a given range.

Q3. What is the difference between discrete and continuous distributions?

Ans:

Discrete Distribution: Defined only for specific, separate values (e.g., binomial, Poisson).

Continuous Distribution: Defined over an interval and includes all values within that range (e.g., normal, uniform).

Q4. What are probability distribution functions (PDF)?

Ans:

For discrete variables, it is called a Probability Mass Function (PMF).

For continuous variables, the Probability Density Function (PDF) gives the relative likelihood of a variable falling within a range of values.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Ans:

PDF shows the likelihood for a specific value or small interval.

CDF gives the probability that the variable takes a value less than or equal to a certain point.

Q8. What is the binomial distribution, and how is it used in probability?

Ans: The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

Q9. What is the Poisson distribution and where is it applied?

Ans: The Poisson distribution models the number of events occurring in a fixed interval of time or space, given the events happen at a constant average rate and independently of the time since the last event.

Q10. What is a continuous uniform distribution?

Ans: A continuous uniform distribution has constant probability density over a specific interval. Every value within the interval is equally likely.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans:
The CLT states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution. It justifies using normal-based methods in many real-world problems.

Q14. How does the Central Limit Theorem relate to the normal distribution?

Ans:
Thanks to the CLT, we can assume that the sample means follow a normal distribution for large sample sizes, even if the original data is not normally distributed.

Q15. What is the application of Z statistics in hypothesis testing?

Ans:
Z statistics help determine how far a sample statistic is from the population mean, measured in standard deviations. It's used to assess the significance of test results.

Q17. What are point estimates and interval estimates in statistics?

Ans:

Point Estimate: A single value estimate of a population parameter (e.g., sample mean).

Interval Estimate: A range (like a confidence interval) that is likely to contain the population parameter.

Q18. What is the significance of confidence intervals in statistical analysis?

Ans:
Confidence intervals provide a range of values for the parameter and indicate the uncertainty associated with the estimate. A 95% CI means we are 95% confident the true value lies within the range.

Q19. What is the relationship between a Z-score and a confidence interval?

Ans:
Z-scores determine the boundaries of a confidence interval in a standard normal distribution. For example, a Z-score of ¬±1.96 corresponds to a 95% confidence level.

Q20. How are Z-scores used to compare different distributions?

Ans:
Z-scores standardize values from different distributions, allowing for meaningful comparison by measuring how many standard deviations each value is from its respective mean.

Q22. What is the concept of expected value in a probability distribution?

Ans:
The expected value is the long-run average outcome of a random variable. It is the weighted average of all possible values.

Q23. How does a probability distribution relate to the expected outcome of a random variable?

Ans:
The probability distribution defines the likelihood of all outcomes. The expected value is calculated using the distribution and reflects the average outcome over many repetitions.
