Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Ans:

What is Anomaly Detection?

Anomaly Detection (also called outlier detection) is the process of identifying rare items, events, or patterns in data that deviate significantly from the expected behavior.

Anomalies often indicate critical incidents like fraud, structural defects, network intrusions, or unexpected events.

For example:

Detecting a sudden spike in credit card spending â†’ possible fraud.

Identifying a strange pattern in a sensor reading â†’ possible machine failure.

Types of Anomalies
1. Point Anomalies

A single data instance is anomalous when it is far away from the rest of the data.

Most common type of anomaly.

Example:

In a dataset of peopleâ€™s monthly expenses (mostly around â‚¹20,000â€“â‚¹40,000), one person spends â‚¹2,00,000 in a month.

This one transaction is a point anomaly.

2. Contextual Anomalies (a.k.a. Conditional Anomalies)

A data instance is anomalous only in a specific context.

The same value may be normal in one context but abnormal in another.

Context is usually defined by time, location, or environment.

Example:

Temperature reading of 30Â°C:

Normal in summer (India).

Abnormal in winter (Himalayan region).

So the anomaly depends on the context.

3. Collective Anomalies

A group of related data instances together is anomalous, even if individual points are not.

Usually detected in time-series data or sequential datasets.

Example:

A sudden sequence of high network traffic requests within a short time (like a DDoS attack).

Each individual request may look normal, but the collective pattern is unusual.

Question 3: What are the key components of a Time Series? Explain each with one
example.

Ans:

Key Components of a Time Series

A time series is a sequence of data points collected or recorded at specific time intervals (daily, monthly, yearly, etc.). To analyze or forecast time series data, we usually break it down into four main components:

1. Trend (T)

Definition: The long-term upward or downward movement in the data.

Represents the general direction of the series over a long period.

Example:

The sales of smartphones have been increasing every year over the last decade due to technological growth.

2. Seasonality (S)

Definition: Repeating short-term cycles that occur at fixed periods (daily, weekly, monthly, yearly).

Caused by calendar-related effects.

Example:

Ice cream sales increase every summer and decrease during winter.

3. Cyclic Component (C)

Definition: Long-term fluctuations that occur irregularly (not fixed like seasonality).

Usually tied to economic or business cycles (expansion, recession, recovery).

Example:

Stock market performance rises during economic growth and falls during recessions.

4. Irregular/Random Component (I)

Definition: Random variations in the data that cannot be explained by trend, seasonality, or cycles.

Caused by unexpected events.

Example:

A sudden drop in airline passengers due to a pandemic or natural disaster.

Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

Ans:

Stationarity in Time Series
Definition

A time series is said to be stationary if its statistical properties (mean, variance, autocovariance) do not change over time.

In other words:

The mean is constant.

The variance is constant.

The covariance between two time points depends only on the lag (difference in time), not on the actual time at which it is measured.

Example

Stationary series: Daily temperature deviations from the average (fluctuate around a constant mean, stable variance).

Non-stationary series: Monthly sales data that shows an upward trend over years (mean is increasing).

Why Stationarity is Important?

Most statistical forecasting methods (like ARIMA) assume stationarity.

Non-stationary data can lead to misleading results in forecasting.

How to Test Stationarity?

Visual Inspection

Plot the data.

If mean/variance changes over time â†’ likely non-stationary.

Summary Statistics

Divide data into different time windows and compare mean/variance.

Statistical Tests

Augmented Dickey-Fuller (ADF) Test

Null Hypothesis (Hâ‚€): Series is non-stationary.

If p-value < 0.05, reject Hâ‚€ â†’ series is stationary.

KPSS Test (Kwiatkowskiâ€“Phillipsâ€“Schmidtâ€“Shin)

Null Hypothesis (Hâ‚€): Series is stationary.

If p-value < 0.05, reject Hâ‚€ â†’ series is non-stationary.

```python
'''Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components
(Include your Python code and output in the code box below.)
'''
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset (comes with statsmodels or can be loaded from CSV)
# If you have CSV: df = pd.read_csv("AirPassengers.csv", parse_dates=['Month'], index_col='Month')

# Built-in dataset from statsmodels
import statsmodels.api as sm
data = sm.datasets.get_rdataset("AirPassengers").data

# Convert to datetime index
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)
ts = data['value']

# Plot the original series
plt.figure(figsize=(10,5))
plt.plot(ts, label='AirPassengers Data')
plt.title("Original AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(ts, model='multiplicative')  # can also use 'additive'

# Plot decomposition
fig = decomposition.plot()
fig.set_size_inches(10,8)
plt.show()
```

```python
'''Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)'''
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Example dataset (NYC Taxi Fare: fare vs distance)
# In practice, you would load with: df = pd.read_csv("nyc_taxi_fare.csv")
# For demo, let's create synthetic data
np.random.seed(42)
n_samples = 200
distance = np.random.normal(5, 2, n_samples)   # average trip distance ~ 5 miles
fare = distance * 3 + np.random.normal(0, 2, n_samples)  # approx $3 per mile

# Add some anomalies
distance = np.append(distance, [20, 25, 30])  # unusually long trips
fare = np.append(fare, [10, 15, 50])          # suspicious fares

df = pd.DataFrame({'distance': distance, 'fare': fare})

# Apply Isolation Forest
clf = IsolationForest(contamination=0.05, random_state=42)
df['anomaly'] = clf.fit_predict(df[['distance', 'fare']])

# -1 = anomaly, 1 = normal
outliers = df[df['anomaly'] == -1]
normal = df[df['anomaly'] == 1]

# Visualization
plt.figure(figsize=(8,6))
plt.scatter(normal['distance'], normal['fare'], c='blue', label='Normal')
plt.scatter(outliers['distance'], outliers['fare'], c='red', label='Anomaly', marker='x', s=100)
plt.title("Isolation Forest - NYC Taxi Fare Anomaly Detection")
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.legend()
plt.show()
```

```python
'''Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)
'''
# SARIMA on AirPassengers â€” train + 12-month forecast
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -------------------------
# 1. Load AirPassengers data
# -------------------------
# statsmodels has the R dataset "AirPassengers"
rdata = sm.datasets.get_rdataset("AirPassengers").data
# rdata has columns 'time' and 'value' in this R dataset
# create a proper datetime index (monthly) from 1949-01 onwards
ts = pd.Series(rdata['value'].values,
               index=pd.date_range(start='1949-01-01', periods=len(rdata), freq='M'),
               name='passengers')

# Quick plot of the original series
plt.figure(figsize=(10,4))
plt.plot(ts, label='Observed')
plt.title('AirPassengers â€” Observed (1949-1960)')
plt.ylabel('Passengers')
plt.xlabel('Year')
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------------------
# 2. Fit a SARIMA model (SARIMAX in sm)
#    Common starting config: (p,d,q)=(1,1,1), (P,D,Q,s)=(1,1,1,12)
# ----------------------------------------
order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12)

model = SARIMAX(ts, order=order, seasonal_order=seasonal_order,
                enforce_stationarity=False, enforce_invertibility=False)
results = model.fit(disp=False)   # disp=False avoids printing optimization details

print("SARIMA model fitted. AIC:", results.aic)

# ----------------------------------------
# 3. Forecast next 12 months
# ----------------------------------------
n_steps = 12
forecast_res = results.get_forecast(steps=n_steps)
forecast = forecast_res.predicted_mean
conf_int = forecast_res.conf_int(alpha=0.05)   # 95% CI

# ----------------------------------------
# 4. Plot observed + forecast + CI
# ----------------------------------------
plt.figure(figsize=(11,5))
plt.plot(ts, label='Observed')
plt.plot(forecast.index, forecast, label='Forecast (next 12 months)', marker='o')
plt.fill_between(conf_int.index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1],
                 color='lightgray', alpha=0.5, label='95% CI')
plt.title('AirPassengers â€” SARIMA Forecast (12 months)')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------------------
# 5. Print forecast table (optional)
# ----------------------------------------
fc_table = pd.DataFrame({
    'forecast': forecast,
    'ci_lower': conf_int.iloc[:, 0],
    'ci_upper': conf_int.iloc[:, 1]
})
print("\nNext 12 months forecast:\n")
print(fc_table.round(2))
```

```python
'''Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)'''
# Local Outlier Factor (LOF) - Anomaly Detection Example
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# -----------------------------
# 1. Create synthetic dataset
# -----------------------------
np.random.seed(42)
X_inliers = 0.3 * np.random.randn(200, 2) + 2   # normal cluster near (2,2)
X_inliers = np.r_[X_inliers, 0.3 * np.random.randn(200, 2) - 2]  # another cluster near (-2,-2)

# Add outliers
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))

# Combine dataset
X = np.r_[X_inliers, X_outliers]
df = pd.DataFrame(X, columns=['x1', 'x2'])

# -----------------------------
# 2. Apply Local Outlier Factor
# -----------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)   # -1 = anomaly, 1 = normal

df['anomaly'] = y_pred

# -----------------------------
# 3. Visualization
# -----------------------------
plt.figure(figsize=(8,6))
plt.scatter(df[df['anomaly']==1]['x1'], df[df['anomaly']==1]['x2'],
            c='blue', label='Normal', alpha=0.6)
plt.scatter(df[df['anomaly']==-1]['x1'], df[df['anomaly']==-1]['x2'],
            c='red', marker='x', s=100, label='Anomaly')
plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Question 10: You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:

â— How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?

â— Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

â— How would you validate and monitor the performance over time?

â— How would this solution help business decisions or operations?
(Include your Python code and output in the code box below.)

Ans:

 Real-Time Power Grid Monitoring Workflow
1. Anomaly Detection

Choice: Isolation Forest (suitable for streaming numeric data, scalable, less sensitive to noise).

Why not DBSCAN/LOF?

DBSCAN is computationally heavy for streaming.

LOF works well but not ideal for large real-time data streams.

How:

Train Isolation Forest on historical energy usage data.

Continuously score incoming 15-min data points to flag spikes/drops.

2. Forecasting

Choice: SARIMAX (Seasonal ARIMA with exogenous variables).

Reason:

Captures seasonality (daily, weekly patterns).

Incorporates external features like weather, region, holidays.

For short-term prediction (next few hours or next day), SARIMAX outperforms plain ARIMA/SARIMA.

3. Validation & Monitoring

Use rolling/expanding window evaluation:

Train on historical â†’ predict next few steps â†’ compare with actuals.

Metrics: MAE, RMSE, MAPE for forecasts.

Track anomaly detection precision/recall against labeled anomaly events.

Deploy a dashboard to monitor forecasts vs actual usage in real-time.

4. Business Impact

Benefits:

Predicts demand peaks â†’ optimize power generation & reduce costs.

Detects abnormal drops/spikes â†’ prevent grid failures, identify theft/faults.

Improves load balancing â†’ reduces blackouts.

Better planning for maintenance and renewable energy integration.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ------------------------
# 1. Simulated Energy Dataset
# ------------------------
np.random.seed(42)
date_rng = pd.date_range(start='2023-01-01', end='2023-03-01', freq='15min')
energy_usage = 200 + 20*np.sin(np.linspace(0, 20, len(date_rng))) + np.random.normal(0, 5, len(date_rng))

# Add anomalies
energy_usage[500:505] += 100  # spike
energy_usage[1500:1505] -= 80 # drop

df = pd.DataFrame({"timestamp": date_rng, "energy": energy_usage})
df.set_index("timestamp", inplace=True)

# ------------------------
# 2. Anomaly Detection (Isolation Forest)
# ------------------------
iso = IsolationForest(contamination=0.01, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy"]])

# Mark anomalies
anomalies = df[df["anomaly"] == -1]

plt.figure(figsize=(12,6))
plt.plot(df.index, df["energy"], label="Energy Usage")
plt.scatter(anomalies.index, anomalies["energy"], color="red", label="Anomalies")
plt.title("Energy Usage with Anomaly Detection")
plt.legend()
plt.show()

# ------------------------
# 3. Forecasting with SARIMAX
# ------------------------
train = df["energy"][:-200]
test = df["energy"][-200:]

# Fit SARIMAX (p,d,q)(P,D,Q,s) with daily seasonality (96 intervals per day)
model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,96))
results = model.fit(disp=False)

forecast = results.forecast(steps=len(test))

plt.figure(figsize=(12,6))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Test", color="orange")
plt.plot(test.index, forecast, label="Forecast", color="green")
plt.title("SARIMAX Forecast for Energy Demand")
plt.legend()
plt.show()

# ------------------------
# 4. Validation
# ------------------------
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}")
```
